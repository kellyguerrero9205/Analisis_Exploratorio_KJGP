2. Análisis univariado

```python
# Estadísticas descriptivas
df.describe()
```

```python
# Histograma de variable Tamaño de empresa
import plotly.express as px

# Contar la frecuencia de cada categoría
conteo_categorias = df['Desc.TamEmpresa'].value_counts().reset_index()
conteo_categorias.columns = ['Desc.TamEmpresa', 'Frecuencia']
# Crear el gráfico de barras con colores diferentes para cada barra
fig = px.bar(
    conteo_categorias,
    x='Desc.TamEmpresa',
    y='Frecuencia',
    color='Desc.TamEmpresa',
    color_discrete_sequence=px.colors.sequential.Blugrn,
    title='Distribución de Tamaño de Empresas de Proveedores'
)

# Ajustar el tamaño del gráfico
fig.update_layout(
    xaxis_title='Tamaño de la Empresa',
    yaxis_title='Frecuencia',
    xaxis_tickangle=-45,
    width=700,  # Ancho del gráfico en píxeles
    height=400  # Alto del gráfico en píxeles
)

# Mostrar el gráfico interactivo
fig.update_layout(
    xaxis_title='Tamaño de la Empresa',
    yaxis_title='Frecuencia',
    xaxis_tickangle=0
)

fig.show()
```

```python
# Histograma de frecuencia de país de origen de proveedor
import plotly.express as px

# Contar la frecuencia de cada categoría
conteo_categorias = df['Pais'].value_counts().reset_index()
conteo_categorias.columns = ['Pais', 'Frecuencia']
# Crear el gráfico de barras con colores diferentes para cada barra
fig = px.bar(
    conteo_categorias,
    x='Pais',
    y='Frecuencia',
    color='Pais',
    color_discrete_sequence=px.colors.sequential.Blugrn,
    title='Cantidad de compras por pais'
)

# Ajustar el tamaño del gráfico
fig.update_layout(
    xaxis_title='Pais de compra',
    yaxis_title='Frecuencia',
    xaxis_tickangle=-45,
    width=1000,
    height=400
)

# Mostrar el gráfico interactivo
fig.update_layout(
    xaxis_title='Pais de compra',
    yaxis_title='Frecuencia',
    xaxis_tickangle=90
)

fig.show()
```

3. Análisis multivariado

```python
# Valor comprado por tamaño de empresa
# Agrupar los datos por tamaño de empresa y sumar el Importe COP
df_grouped = df.groupby('Desc.TamEmpresa')['      Importe COP'].sum().reset_index()

# Crear el gráfico de barras
fig = px.bar(
    df_grouped,
    x='Desc.TamEmpresa',
    y='      Importe COP',
    color='Desc.TamEmpresa',
    color_discrete_sequence=px.colors.qualitative.Pastel2,
    title='Valor comprado por Tamaño de Empresa'
)

# Ajustar el tamaño del gráfico
fig.update_layout(
    xaxis_title='Tamaño de la Empresa',
    yaxis_title='      Importe COP',
    xaxis_tickangle=0,
    width=900,  # Ancho del gráfico en píxeles
    height=500,  # Alto del gráfico en píxeles
    margin=dict(l=40, r=40, t=40, b=40)  # Ajustar los márgenes (izquierda, derecha, arriba, abajo)
)

# Mostrar el gráfico interactivo
fig.show()
```

```python
# Valor comprado por pais
# Agrupar los datos por tamaño de empresa y sumar el Importe COP
df_grouped = df.groupby('Pais')['      Importe COP'].sum().reset_index()

# Crear el gráfico de barras
fig = px.bar(
    df_grouped,
    x='Pais',
    y='      Importe COP',
    color='Pais',
    color_discrete_sequence=px.colors.qualitative.Prism,
    title='Valor comprado por pais'
)

# Ajustar el tamaño del gráfico
fig.update_layout(
    xaxis_title='Pais',
    yaxis_title='      Importe COP',
    xaxis_tickangle=90,
    width=900,  # Ancho del gráfico en píxeles
    height=500,  # Alto del gráfico en píxeles
    margin=dict(l=40, r=40, t=40, b=40)  # Ajustar los márgenes (izquierda, derecha, arriba, abajo)
)

# Mostrar el gráfico interactivo
fig.show()
```

4. Limpieza y preparación de datos

```python
# Imputamos los valores vacios de la columna tipo de material

df['Tipo Mat.'].fillna('Servicio', inplace=True)
df.head()
```

```python
# Valor comprado por grupo de tipo de material
# Agrupar los datos por tamaño de empresa y sumar el Importe COP
df_grouped = df.groupby('Tipo Mat.')['      Importe COP'].sum().reset_index()

# Crear el gráfico de barras
fig = px.bar(
    df_grouped,
    x='Tipo Mat.',
    y='      Importe COP',
    color='Tipo Mat.',
    color_discrete_sequence=px.colors.qualitative.Prism,
    title='Valor comprado por tipo de material'
)

# Ajustar el tamaño del gráfico
fig.update_layout(
    xaxis_title='Tipo Mat.',
    yaxis_title='      Importe COP',
    xaxis_tickangle=90,
    width=900,  # Ancho del gráfico en píxeles
    height=500,  # Alto del gráfico en píxeles
    margin=dict(l=40, r=40, t=40, b=40)  # Ajustar los márgenes (izquierda, derecha, arriba, abajo)
)

# Mostrar el gráfico interactivo
fig.show()
```

```python
# ERSA - repuestos
# HAWA - comercializados
# HIBE - Aceites lubricantes y grasas
# MODE - Dotación
# ROH - Materias primas
# UNBW - No valorados
# VERP - Material de empaque
# ZHAW - Cafetería y enfermería
# ZHIB - MP Auxiliares
# ZREP - Repuestos recuperados
```

```python
#Trabajamos la columna Condición de pago para que sea numerica
from sklearn.preprocessing import LabelEncoder

# Crear una instancia de LabelEncoder
label_encoder = LabelEncoder()

# Aplicar Label Encoding a la columna 'Cond.Pago Ped'
df['Cond.Pago Ped Encoded'] = label_encoder.fit_transform(df['Cond.Pago Ped.1'].astype(str))

# Mostrar el resultado
df.head()
```

```python
df_count = df.groupby(['Desc.TamEmpresa', 'Cond.Pago Ped.1']).size().reset_index(name='Count')

# Crear el gráfico de barras agrupadas
fig = px.bar(
    df_count,
    x='Desc.TamEmpresa',
    y='Count',
    color='Cond.Pago Ped.1',
    title='Frecuencia de Condición de Pago por Tamaño de Empresa',
    labels={'Desc.TamEmpresa': 'Tamaño de Empresa', 'Count': 'Frecuencia'},
    barmode='group'
)

# Ajustar el tamaño del gráfico
fig.update_layout(
    width=1200,
    height=500,
    xaxis_title='Tamaño de Empresa',
    yaxis_title='Frecuencia',
    xaxis_tickangle=-45,
    margin=dict(l=40, r=40, t=40, b=40)
)

# Mostrar el gráfico interactivo
fig.show()
```

```python
# Creamos un df agrupado por material y proveedor
df_grouped = df.groupby(['Cod.Material', 'Material', 'Cod.Proveedor', 'Proveedor', 'Tipo Mat.'], as_index=False)['      Importe COP'].sum()
df_sorted = df_grouped.sort_values(by='      Importe COP', ascending=False)

# Mostrar el DataFrame resultante
df_grouped.head(10)
```

```python
# Contar el número de proveedores únicos por Código de Material
unique_providers_per_material = df_grouped.groupby('Cod.Material')['Cod.Proveedor'].nunique()

# Filtrar los materiales que tienen exactamente un único proveedor
single_provider_materials = unique_providers_per_material[unique_providers_per_material == 1].index

# Filtramos solo para materias primas y comercializados
df_filtered = df_grouped[df_grouped['Tipo Mat.'].isin(['ROH', 'HAWA'])]

# Obtener el DataFrame con solo los materiales que tienen un único proveedor
df_single_provider_materials = df_filtered[df_filtered['Cod.Material'].isin(single_provider_materials)]

# Mostrar el DataFrame resultante
df_single_provider_materials.head(10)
```

```python
# Ordenar el DataFrame por Importe COP en orden descendente
df_sorted = df_single_provider_materials.sort_values(by='      Importe COP', ascending=False)

# Seleccionar el top 10 por Importe COP
df_top10 = df_sorted.head(10)

df_top10.head(10)
```

```python
# Crear el gráfico de barras interactivo con una paleta de colores válida
fig = px.bar(
    df_top10,
    x='Material',
    y='      Importe COP',
    color='      Importe COP',
    text='Proveedor',
    color_continuous_scale='viridis',
    labels={'Material': 'Código de Material', '      Importe COP': 'Importe COP'},
    title='Top 10 de para Materiales con un Único Proveedor'
)

# Ajustar el diseño del gráfico
fig.update_layout(
    xaxis_title='Material',
    yaxis_title='Importe COP',
    xaxis_tickangle=-45
)

# Ajustar las etiquetas de texto
fig.update_traces(texttemplate='%{text}', textposition='outside')

# Mostrar el gráfico
fig.show()
```
