```python
# Para cargue de datos desde el directorio local
from google.colab import files
uploaded = files.upload()
```

El archivo elegido corresponde a datos del indice de costo de vida por país en el 2024, tomando como referencia la ciudad de NY.

```python
# Validamos el cargue
for filename in uploaded.keys():
    print(f"Archivo subido: {filename}")
```

```python
# Generamos el dataframe y lo visualizamos
import pandas as pd
medallas_pais = list(uploaded.keys())[0]
df = pd.read_csv(medallas_pais)
print(df.head())
```

```python
# Validamos el tamaño del dataframe
num_rows, num_cols = df.shape
print(f"Número de filas: {num_rows}")
print(f"Número de columnas: {num_cols}")
```

El dataframe contiene información en 8 columnas y 121 filas, es decir datos de 121 países.

```python
# Visualizamos el tipo de dato de cada atributo
print(df.dtypes)
```

Los datos recolectados son de tipo número excepto el país

```python
df.describe()
```

Enfocandonos en el indice de costo de vida podemos afirmar que se recolectaron 121 datos, que en promedio el costo de vida de estas 121 ciudades es 60% más barato que NY, que el dato mínimo nos informa que hay al menos una ciudad con un costo de vida 81,2% más barato que NY, y el dato máximo nos informa que hay al menos una ciudad con un costo de vida 21% mayor que NY. También nos informa la desviación sts de 16,4 que los datos estan concentrados y cercanos a la media.

```python
df.describe(include = "all")
```

```python
import altair as alt
chart = alt.Chart(df).mark_circle().encode(x='Cost of Living Index', y='Rent Index')
chart
```

Este gráfico de dispersión nos afirma que los datos estan mayormente concentrados. Adicional, nos muestra una correlación positiva frente a indice de renta, lo que se traduce que a mayor indice de rente, mayor indice de vida.

```python
missing_data = df.isnull()
missing_data.head()
```

Esta consulta nos permite visualizar que al menos en las 4 primeras filas no se tienen datos nulos o perdidos

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

Esta consulta nos confirma que para ningún tipo de dato, existen datos nulos o perdidos
